## Was ist AlphaGo?

- AlphaGo ist ein Computerprogramm, das das Brettspiel **Go** spielt 
- wurde von der Firma **DeepMind** entwickelt
- AlphaGo kombiniert Techniken des **maschinellen Lernens** und der **Traversierung**

## Was ist Go?

- ist ein strategisches Brettspiel für zwei Spieler
- stammt ursprünglich aus dem antiken China 
- hat im Laufe der Geschichte eine besondere Prägung in Japan, Korea und Taiwan erhalten 
- erst seit dem 20. Jahrhundert fand Go auch Verbreitung außerhalb Ostasiens 
- heute befinden sich die meisten Go-Spieler in Ostasien

### Wie wird Go gespielt?

- Spieler setzen abwechselnd jeweils einen eigenen Stein auf die Schnittpunkte der Linien des Brettes
- Man kann gegnerische Steine und Steingruppen schlagen, indem man sie rundum einschließt
- Am Ende gewinnt der Spieler, der den größeren Teil des Brettes kontrolliert

## Netze von AlphaGo

AlphaGo verwendet zusätzlich zu Monte-Carlo-Methoden Lernmethoden für **tiefe neuronale Netzwerke**.<br>
Dabei werden **zwei** Kategorien von neuronalen Netzen und eine Baumsuche eingesetzt:

- **policy network** („Regelnetzwerk“) wird zur Bestimmung von Zugkandidaten mit großen Mengen von Partien sowohl durch *überwachtes Lernen* (engl. supervised learning) konditioniert als auch durch *bestärkendes Lernen* (engl. reinforcement learning) trainiert
- **value network** („Bewertungsnetzwerk“) dient der Bewertung von Positionen und wird durch *bestärkendes Lernen* eingestellt.
- **Monte-Carlo-Baumsuche** rechnet die Varianten durch. Alle drei Komponenten werden in dieser Baumsuche kombiniert.

### Lernansatz von AlphaGo

- Ansatz auf andere Anwendungsgebiete übertragbar
- Programm lernt durch Analyse einer Datenbank von 30 Millionen Zügen, den Zug eines Menschen „vorherzusagen“
- gelingt zu 56 %
- Partie muss für eine Bewertung eines Zuges nicht zuende gespielt werden
- in Praxis werden für die stärkste Version von AlphaGo zusätzlich auch noch Bewertungen nach dem Monte-Carlo-Verfahren vorgenommen
---
#### Praxisbeispiele:
- bei Partien gegen Fan Hui lief die verteilte Variante von AlphaGo auf einem Rechnerverbund mit insgesamt 1202 CPUs und 178 GPUs und nutzte 40 Such-Threads (search threads)
- bei späteren Matches gegen Lee Sedol wurden 1920 CPUs und 280 GPUs verwendet
- um während der Lernphase benötigte massive Rechenleistung bereitzustellen, wurden die Google Cloud Platform und TensorFlow Processing Units (ASICs für die Software-Sammlung TensorFlow) eingesetzt

- Lee gilt als einer der derzeit besten Spieler der Welt
