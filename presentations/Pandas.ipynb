## Series

```python
quandl.ApiConfig.api_key = ''
apple = quandl.get('WIKI/AAPL')
ms = quandl.get('WIKI/MSFT')
```

```python
ms.head()
```

### Dataframe vs Series

```python
type(ms)
```

```python
type(ms['Adj. Close'])
```

```python
type(ms[['Adj. Close']])
```

### Using Series

```python
ms['Adj. Close'].plot()
```

```python
ms.loc['2018']['Adj. Close'].plot()
```

```python
ms.loc['2018-03']['Adj. Close'].plot()
```

```python
ms.loc['2018-01-01' : '2018-03-31']['Adj. Close'].plot()
```

```python
ms_price = ms[['Adj. Close']].copy()
ms_price.rename(columns = {'Adj. Close' : 'MSFT'}, inplace = True)
ms_price.head()
```

```python
apple_price = apple[['Adj. Close']].copy()
apple_price.rename(columns = {'Adj. Close' : 'AAPL'}, inplace = True)
apple_price.head()
```

```python
both_stocks = ms_price.join(apple_price, how='inner')
```

```python
both_stocks.head()
```

```python
both_stocks.plot()
```

```python
both_stocks.loc['2017'].plot()
```

```python
both_stocks.loc['2017'].rolling(window = 30).mean().plot()
```
